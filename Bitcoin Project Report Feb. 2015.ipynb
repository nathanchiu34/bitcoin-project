Let's split up them up into years:

```python
# Split up the Quandl data into 2013 data
df_bitcoin13 = pd.DataFrame() 
for i in df_bitcoin.index:
	if i.year == 2013: 
		df_bitcoin13 = df_bitcoin13.append(df_bitcoin.loc[str(i)])#ignore_index=True)

# Split up the Quandl data into 2014 data
df_bitcoin14 = pd.DataFrame()
for i in df_bitcoin.index:
	if i.year == 2014:
		df_bitcoin14 = df_bitcoin14.append(df_bitcoin.loc[str(i)])

# Split up Quandl SP 500 data into 2013 dat
df_sp500_13 = pd.DataFrame()
for i in df_sp500.index:
	if i.year == 2013:
		df_sp500_13 = df_sp500_13.append(df_sp500.loc[str(i)])
        
# Split up Quandl SP 500 data into 2013 dat
df_sp500_14 = pd.DataFrame()
for i in df_sp500.index:
	if i.year == 2014:
		df_sp500_14 = df_sp500_14.append(df_sp500.loc[str(i)])  
        
        
```

Let's take a look at the 2014 data for both Bitcoin and S&P 500 volume statistics. I created a function to display in scatterplots the data.

```python
figsize(8,8)

def compare_volume():
	#Volume 13 and 14
	vol_SP500_13 = np.array(df_sp500_13['Volume'])
	vol_Bitcoin13 = np.array(df_bitcoin13['Volume (BTC)'])
	vol_Bitcoin14 = np.array(df_bitcoin14['Volume (BTC)'])
	plt.scatter( df_bitcoin13.index, vol_Bitcoin13, marker = 'x', color= 'blue')
	plt.scatter( df_bitcoin14.index, vol_Bitcoin14, marker = 'o', color ='green')
	plt.title("Volume of Bitcoin",fontsize = 14)
    #Adjusted Volume 13 and 14, ( Volume / Average Volume )    

def compare_adjvolume():
    adj_vol_BC13 = np.array(df_bitcoin13['Volume (BTC)'] / df_bitcoin13['Volume (BTC)'].mean())
    adj_vol_BC14 = np.array(df_bitcoin14['Volume (BTC)'] / df_bitcoin14['Volume (BTC)'].mean())
    plt.scatter( df_bitcoin13.index, adj_vol_BC13, marker = 'x', color = 'blue')
    plt.scatter( df_bitcoin14.index, adj_vol_BC14, marker = 'o', color = 'green')
    plt.title("Adjusted Volume of Bitcoin", fontsize = 14)

subplot(1,2,1)
compare_volume()
subplot(1,2,2)
compare_adjvolume()
```
